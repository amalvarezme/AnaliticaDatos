# APIs

1. ¿Qué Son las APIs?
    - De sus siglas en inglés **Application Programming Interface**
    - Imagine que está trabajando en un proyecto con su compañero, usted trabaja en PYTHON y el en JAVA, una de las primeras opciones que se les puede ocurrir es escribir nuevamente el código en cualquiera de los dos lenguajes, pero claro se requiere mucho trabajo y más tiempo. 
  - Las APIs fueron diseñadas para ahorrar este trabajo y crear un "puente" entre los dos código y obtener el resultado esperado.
     > Una API en simples palabras, se define como un "contrato" entre dos softwares donde se define el usuario de la entrada en cierto formato predeterminado, extendiendo su funcionalidad y dando el resultado al software del usuario. 
     [Analytics Vidhya](https://www.analyticsvidhya.com/blog/2016/11/an-introduction-to-apis-application-programming-interfaces-5-apis-a-data-scientist-must-know/)
     
    - En general las APIs son muy parecidas a las aplicaciones de web, pero en lugar de mostrar un paǵina HTML elegante, tienden a retornar datos en un formato estándar *data-exchange*, formatos como:
         - JSON.
         - XML
         
         
2. __APIs en Python__.

    - En Python podemos crear nuestras APIs con cualquiera de las siguientes herramientas:
    
        - [Hug](http://www.hug.rest/)
        - [Flask](https://flask.palletsprojects.com/en/1.1.x/)
        - [Django](https://tutorial.djangogirls.org/es/django/)  
        
    - Nosotros vamos a trabajar con **Flask**, flask es un marco de desarrollo de servicios web en python


__El Archivo model_pkl.py__

``` python
import pandas as pd


# Load the dataset in a dataframe object and include only four features as mentioned
url = "http://s3.amazonaws.com/assets.datacamp.com/course/Kaggle/train.csv"
df = pd.read_csv(url)
include = ['Age', 'Sex', 'Embarked', 'Survived'] # Only four features
df_ = df[include]

# Data Preprocessing
categoricals = []
for col, col_type in df_.dtypes.iteritems():
    if col_type == 'O':
        categoricals.append(col)
    else:
        df_[col].fillna(0, inplace=True)

df_ohe = pd.get_dummies(df_, columns=categoricals, dummy_na=True)

# Logistic Regression classifier
from sklearn.linear_model import LogisticRegression
dependent_variable = 'Survived'
x = df_ohe[df_ohe.columns.difference([dependent_variable])]
y = df_ohe[dependent_variable]
lr = LogisticRegression()
lr.fit(x, y)

# Save your model
from sklearn.externals import joblib
joblib.dump(lr, 'models/model.pkl')
print("Model dumped!")

# Load the model that you just saved
lr = joblib.load('model.pkl')

# Saving the data columns from training
model_columns = list(x.columns)
joblib.dump(model_columns, 'models/model_columns.pkl')
print("Models columns dumped!")

```
